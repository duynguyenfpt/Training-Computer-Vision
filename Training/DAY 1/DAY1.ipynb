# Bài tập 3: Drawing Functions in OpenCV

Sẽ học các function:

- cv2.line() : Vẽ đường thẳng lên ảnh.
- cv2.circle(): Vẽ một đường tròn lên ảnh.
- cv2.rectangle(): Vẽ một hình chữ nhật lên ảnh.
- cv2.ellipse(): Vẽ đường cong dựa vào góc truyền vào esclipe lên ảnh.
- cv2.putText(): Vẽ một input text lên ảnh.

```python
img = cv2.imread(rel_path[0])
```

```python
# Draw a diagonal blue line with thickness of 5 px
imgTmp = img.copy()
imgLine = cv2.line(imgTmp,(0,0),(100,100),(255,0,0),5)
plt.imshow(imgLine)
```

```python
# Draw a green ractangle line with thickness of 5 px
imgTmp = img.copy()
imgRec = cv2.rectangle(imgTmp,(0,0),(100,100),(0,255,0),3)
plt.imshow(imgRec)
```

```python
# Draw a diagonal blue circle with radius of 60px
imgTmp = img.copy()
imgCircle = cv2.circle(imgTmp,(100,100), 60, (0,0,255), -1)
plt.imshow(imgTmp)
```

```python
imgTmp = img.copy()
imgElipse = cv2.ellipse(imgTmp,(100,100),(30,20),0,0,180,255,-1)
plt.imshow(imgElipse)
```

```python
imgTmp = img.copy()
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
imgPolygon = cv2.polylines(imgTmp,[pts],True,(0,255,255))
plt.imshow(imgPolygon)
```

```python
imgTmp = img.copy()
font = cv2.FONT_HERSHEY_SIMPLEX
imgText = cv2.putText(imgTmp,'Weboo',(50,50), font, 1,(255,255,255),1,cv2.LINE_AA)
plt.imshow(imgText)

cv2.imshow("hihi",imgTmp)
cv2.waitKey(0)
cv2.destroyWindow("hihi")
```

# Bài tập 4: Làm quen với thư viện Numpy và OpenCV

Numpy là một trong những thư viện hay được sử dụng nhất trong Python, thư viện giúp chúng ta dễ dàng thực hiện các phép toán, các phép biến đổi trên ma trận nhiều chiều. 

OpenCV là một trong những framework tốt nhất hiện tại dùng trong xử lý ảnh với việc hỗ trợ rất nhiều hàm, phép biến đổi quan trọng, giúp cho việc lập trình các mô hình trở nên dễ dàng hơn rất nhiều. OpenCV chủ yếu xử lý trên các ảnh và video, mà ảnh và video lại được lưu dưới dạng các ma trận số có thể xử lý bằng thư viện numpy. Do đó việc thành thạo 2 thư viện này là yêu cầu bắt buộc.

### Bài tập 4.1: Hiển thị các kênh màu của ảnh
Ở bài tập này ta sẽ hiển thị lần lượt các kênh màu B, G, R của ảnh ba chiều. Cách làm: Gán các giá trị cường độ màu tại các pixel của kênh màu không quan tâm bằng 0. Ví dụ: Nếu cần hiển thị kênh màu xanh lam (B) thì gán các giá trị cường độ màu G, R bằng 0:

```python
## HOME WORK

img = cv2.imread(rel_path[1])

b_channel = None
g_channel = None
r_channel = None

#### YOUR CODE HERE ####
b_channel = img.copy()
b_channel[:, :, 1:] = 0

g_channel = img.copy()
g_channel[:, :, [0, 2]] = 0

r_channel = img.copy()
r_channel[:, :, [0, 1]] = 0

#### END YOUR CODE #####

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(b_channel, cv2.COLOR_BGR2RGB))
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(g_channel, cv2.COLOR_BGR2RGB))
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(r_channel, cv2.COLOR_BGR2RGB))
plt.show()
```

### Bài tập 2.2: Xây dựng vector histogram của ảnh xám

Bài tập này giúp bạn làm quen với thao tác trên mảng numpy. Bạn cần lập trình tính giá trị vector histogram, đây là vector gồm 256 phần tử, phần tử thứ i là số lần xuất hiện của giá trị cường độ i trong ảnh xám 'gray':

```python
## HOME WORK

img = cv2.imread(rel_path[2])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape)
histogram = np.zeros((256, ))
print(histogram.shape)

#### YOUR CODE HERE ####
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        histogram[gray[i][j]] += 1
#### END YOUR CODE #####

plt.plot(histogram)
plt.show()
```
